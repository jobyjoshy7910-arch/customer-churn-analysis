```python
import pandas as pd
import numpy as np

df = pd.read_csv('telco_churn.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
df.head()
```

```python
print("Missing values:\n", df.isnull().sum())
print("\nChurn distribution:")
print(df['Churn'].value_counts())
print(f"\nChurn rate: {(df['Churn']=='Yes').mean()*100:.2f}%")
```

```python
problem_statement = """
PROJECT: Telco Customer Churn Analysis

BUSINESS PROBLEM:
A telecom company is losing customers to competitors.
The business team needs to understand:
  1. Which customer segments have the highest churn rate?
  2. What contract types, tenure, and services drive churn?
  3. How much revenue is at risk from churning customers?
  4. What interventions can reduce churn rate?

DATASET:
  Source   : IBM Sample Data (Kaggle)
  Size     : 7,043 customers, 21 features
  Churn Rate: ~26.5% (1,869 churned out of 7,043)

SUCCESS METRICS:
  - Identify top 3 churn drivers from EDA
  - SQL queries showing revenue at risk by segment
  - Dashboard with churn rate by contract, tenure, charges
  - Business recommendation with estimated retention value
"""
print(problem_statement)

with open('problem_statement.txt', 'w') as f:
    f.write(problem_statement)
print("Saved!")
```

```python
print("=== KEY BUSINESS METRICS ===\n")

total = len(df)
churned = (df['Churn'] == 'Yes').sum()
retained = (df['Churn'] == 'No').sum()
avg_monthly = df['MonthlyCharges'].mean()
churned_revenue = df[df['Churn']=='Yes']['MonthlyCharges'].sum()

print(f"Total Customers      : {total:,}")
print(f"Churned Customers    : {churned:,} ({churned/total*100:.1f}%)")
print(f"Retained Customers   : {retained:,} ({retained/total*100:.1f}%)")
print(f"Avg Monthly Charge   : ${avg_monthly:.2f}")
print(f"Monthly Revenue Risk : ${churned_revenue:,.2f}")
print(f"Annual Revenue Risk  : ${churned_revenue*12:,.2f}")
```
